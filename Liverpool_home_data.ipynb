### Combining all data together

```python
df=pd.read_csv("./premier-league/season-1011.csv")
files=[file for file in os.listdir("./premier-league")]
#combining the data:-
alldata=pd.DataFrame()
for file in files:
    df=pd.read_csv("./premier-league/"+file)
    alldata=pd.concat([alldata,df])

alldata.to_csv("all_data.csv",index=False)
```

```python
all_data=pd.read_csv("all_data.csv")
all_data
```

```python
nandf=all_data[all_data.isna().any(axis=1)]
nandf.head()
all_data=all_data.dropna(how="all")
all_data
```

```python
all_data
```

### Liverpool Home Record for 2009-19

```python
liverpool=all_data.loc[all_data["HomeTeam"]=="Liverpool"]
```

```python
liverpool
```

```python
liverpool.head()
```

```python
liverpool.to_csv("Liverpool.csv")
```

```python
liverpool["Month"]=liverpool["Date"].str[3:5] 
liverpool["Month"]=liverpool["Month"].astype("int64") 
liverpool.head()
```

```python
liverpool["Year"]=liverpool["Date"].str[6:8] 
liverpool["Year"]=liverpool["Year"].astype("int64") 
liverpool.head()
```

```python
liverpool.rename(columns={"FTHG":"Goals Scored"}, inplace=True)
```

```python
liverpool.head()
```

```python
liverpool.rename(columns={"FTAG":"Goals Conceded"}, inplace=True)
liverpool.rename(columns={"FTR":"Result"}, inplace=True)
liverpool.rename(columns={"HS":"HT Shots"}, inplace=True)
liverpool.rename(columns={"AS":"AT Shots"}, inplace=True)
liverpool.rename(columns={"HST":"HT SOT"}, inplace=True)
liverpool.rename(columns={"AST":"AT SOT"}, inplace=True)
liverpool.rename(columns={"HF":"HT Fouls"}, inplace=True)
liverpool.rename(columns={"AF":"AT Fouls"}, inplace=True)
liverpool.rename(columns={"HC":"HT Corners"}, inplace=True)
liverpool.rename(columns={"AC":"AT Corners"}, inplace=True)
liverpool.rename(columns={"HY":"HT Yellow"}, inplace=True)
liverpool.rename(columns={"AY":"AY Yeloow"}, inplace=True)
liverpool.rename(columns={"HR":"HT Red"}, inplace=True)
liverpool.rename(columns={"AR":"AT Red"}, inplace=True)
```

```python
liverpool.head()
```

```python
liverpool.to_csv("Liverpool_new_data.csv")
```

### Year-wise performance

```python
results=liverpool.groupby("Year").sum()
results
```

### Which year saw the most goals scored while playing at home?

```python
import matplotlib.pyplot as plt    
year=range(9,19)
plt.bar(year,results["Goals Scored"])
plt.xticks(year)
plt.ylabel("Goals Scored",fontsize=15)
plt.xlabel("Year",fontsize=15)
plt.show()
```

### Which year saw the most goals conceded while playing at home?

```python
year=range(9,19)
plt.bar(year,results["Goals Conceded"])
plt.xticks(year)
plt.ylabel("Goals Conceded",fontsize=15)
plt.xlabel("Year",fontsize=15)
plt.show()
```

```python
ref=liverpool.groupby("Referee").sum()
ref
```

### Referee with the most goals in home games

```python
import matplotlib.pyplot as plt 
referee=[r for r, df in liverpool.groupby("Referee")]
plt.bar(referee,ref["Goals Scored"])
plt.xticks(referee,rotation="vertical",size=10)
plt.ylabel("Goals Scored")
plt.xlabel("Referee Name")
plt.show()
```

### Referee with the most Yellow Cards awarded to Liverpool at home

```python
referee=[r for r, df in liverpool.groupby("Referee")]
plt.bar(referee,ref["HT Yellow"])
plt.xticks(referee,rotation="vertical",size=10)
plt.ylabel("Yellow Cards")
plt.xlabel("Referee Name")
plt.show()
```

### Referee with the most Red Cards awarded to Liverpool at home

```python
referee=[r for r, df in liverpool.groupby("Referee")]
plt.bar(referee,ref["HT Red"])
plt.xticks(referee,rotation="vertical",size=10)
plt.ylabel("Red Cards")
plt.xlabel("Referee Name")
plt.show()
```
